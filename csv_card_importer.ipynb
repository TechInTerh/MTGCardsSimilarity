```python
df_sets = pd.read_csv("AllPrintingsCSVFiles/sets.csv")
df_sets["releaseDate"] = pd.to_datetime(df_sets["releaseDate"])
df_sets
```

```python
df_cards = df_cards.merge(df_cards_legal, on="uuid")
df_cards
```

```python
df_cards = df_cards.merge(df_sets, left_on="setCode", right_on="code",suffixes=("","_set"))
df_cards
```

```python
df_cards_legal = df_cards.dropna(subset=["commander"])
df_cards_legal
```

```python
df_cards_legal.dtypes
```

```python
df_cards_legal = df_cards_legal.sort_values("releaseDate").drop_duplicates("name",keep="last") # Keep the last version of the card
df_cards_legal
```
